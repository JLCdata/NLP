````python
def get_tokens(text):
  ### Inicio del c칩digo ###
  # Separacion de palabras por espacios
    v1=text.split()

    # Separacion por puntos, comas y comillas
    v2=[]
    for i in v1:
        if "." in i:
            v2.append(i[0:-1])
            v2.append(i[-1])
        
        elif "'" in i:
            v2.append(i[0:-1])
            v2.append(i[-1])
        
        elif "," in i:
            v2.append(i[0:-1])
            v2.append(i[-1])
        
        else:
            v2.append(i)

    # Separacion por parentesis
    v3=[]
    for i in v2:

        if ("("  in i) and ( ")" in i):
        
            v3.append(i[0])
            v3.append(i[1:-1])
            v3.append(i[-1])
        
        elif "(" in i:
            v3.append(i[0])
            v3.append(i[1:])

        elif ")" in i:
            v3.append(i[0:-1])
            v3.append(i[-1])

        else:
            v3.append(i)

    # Separaci칩n de palabras por guiones
    v4=[]
    for i in v3:
        
        if "-" in i:
            indice=i.find("-")
            v4.append(i[0:indice])
            v4.append(i[indice])
            v4.append(i[indice+1:])
        else:
            v4.append(i)

    return v4
  ### F칤n del c칩digo ###
````

````python
tokens = get_tokens(text)
tokens
````

Ahora dise침e reglas que usted estime convenientes tanto de **Stemming** como de **Stopwords**. Implemente una funci칩n que reciba una lista con los elementos del vocabulario, le aplique sus reglas y devuelva el vocabulario actualizado. 

Primero se normalizan los terminos cambiando todas las letras a min칰sculas. Luego se eliminan stopwords para palabras en Ingles. Finalmente se define la regla de Stemming que cambia los terminos de palabras "es" en "e".    

````python
def pre_processing(vocab): 

  ### Aqu칤 inicia tu c칩digo ###

  # Normalizacion a minusculas
  norm=pd.Series(vocab).apply(lambda x: x.lower())
  
  # Stopwords
  stopwords=[",",".","'","-","(",")","the","is","my","i"]

  # Sin Stopwords
  sin_stopwords=norm[~norm.isin(stopwords)]
  
  # Stemming: Si termina en "es" dejar solamente "e"
  vocab_new=sin_stopwords.apply(lambda x: x[0:-1] if x[-2:]=="es" else x).unique().tolist()
  
  return vocab_new
  ### Aqu칤 termina tu c칩digo ###
````

````python
vocab = pre_processing(vocab)
vocab
````
